#### 05 Define and apply function 'price_label' to subset df

```python
#define function
def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return 'Not enough data'
```

```python
#apply function to df

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

02. IC want periods of time labeled â€œMost orders,â€ â€œAverage orders,â€ and â€œFewest orders.â€ Create a new column containing these labels called â€œbusiest_period_of_day.â€

```python
# frequency of 'order_hour_of_day'

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# create for-loop function to assign busiest day values in new column

result03 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in (10,11,14,15,13,12,16,9):
    result03.append("Most orders")
  elif value in (17,8,18,19,20,7,21,22):
    result03.append("Average orders")
  else:
    result03.append("Fewest orders")
```
