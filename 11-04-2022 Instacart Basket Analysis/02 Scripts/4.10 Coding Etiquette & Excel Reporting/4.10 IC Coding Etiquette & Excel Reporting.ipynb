## 09 Customer Profiles

The marketing and business strategy units at Instacart want to create more-relevant marketing strategies for different products and are, thus, curious about customer profiling in their database. Create a profiling variable based on age, income, certain goods in the “department_id” column, and number of dependents. You might also use the “orders_day_of_the_week” and “order_hour_of_day” columns if you can think of a way they would impact customer profiles. (Hint: As an example, try thinking of what characteristics would lead you to the profile “Single adult” or “Young parent.”)

### A. age_group profile

```python
active['age'].value_counts(dropna=False)
```

```python
# create age_group flag

active.loc[active['age']<=30, 'age_group']='young adult'
active.loc[(active['age']>30)&(active['age']<65), 'age_group']='adult'
active.loc[active['age']>=65, 'age_group']='senior'
active['age_group'].value_counts()
```

### B. age_gender_group profile

```python
# merge age_group column and gender column

active['age_gender_group'] = active[['age_group', 'gender']].agg('-'.join, axis=1)

active['age_gender_group'].value_counts()
```

### C. income_group profile

```python
# create income_group flag

active.loc[active['income']<48500, 'income_group']='lower income'
active.loc[(active['income']>=48500)&(active['income']<=145500), 'income_group']='middle income'
active.loc[active['income']>145500, 'income_group']='upper income'
active['income_group'].value_counts()
```

### D. age_family_status profile

```python
# merge age_group column and family status column

active['age_family_status'] = active[['age_group', 'family_status']].agg('-'.join, axis=1)

active['age_family_status'].value_counts()
```

### E. dependent_group profile

```python
# create dependent_group flag based on no dependents, single dependent, multiple dependents

active.loc[active['number_of_dependants']==0, 'dependent_group']='no dependents'
active.loc[active['number_of_dependants']==1, 'dependent_group']='single dependent'
active.loc[active['number_of_dependants']>1, 'dependent_group']='multiple dependents'
active['dependent_group'].value_counts()
```

### F. order_day_profile

```python
active['orders_day_of_week'].value_counts(dropna=False)
```

```python
# create list of conditions for weekday_shopper vs weekend_shopper

conditions = [
    (active['orders_day_of_week'] ==0),
    (active['orders_day_of_week'] ==6),
    (active['orders_day_of_week'] >0) & (active['orders_day_of_week'] <6)
]
```

```python
# create a list of values to assign

values = ['weekend shopper', 'weekend shopper', 'weekday shopper']
```

```python
active['order_day_profile'] = np.select(conditions, values)
```

```python
active['order_day_profile'].value_counts(dropna=False)
```

## 10 Department Profiles

### A. convert df_dept to data dictionary

```python
df_dept
```

```python
# rename unnamed:0 column to department_id
df_dept.rename(columns = {'Unnamed: 0' : 'department_id'}, inplace = True)
```

```python
df_dept
```

```python
# convert df_dept into data dictionary

zipped=zip(df_dept['department_id'], df_dept['department'])
data_dict=dict(zipped)
data_dict
```

### B. Map department data dict to active data set

```python
# mapping department names to department_id in dataframe
active['department'] = active['department_id'].map(data_dict)
active.head()
```

### C. Create 'alcohol_purchased' department profile



```python
# create column 'alcohol_purchased' 1 if yes in 'department' column, 0 if no
 
active['alcohol_purchased'] = ['Yes' if x == 'alcohol' else 'No' for x in active['department']]
active['alcohol_purchased'] = active.groupby(['user_id'])['alcohol_purchased'].transform(np.max)
active['alcohol_purchased'].value_counts(dropna=False)
```

### D. Create 'dairy_eggs_purchased' department profile

```python
# create column 'dairy_eggs_purchased' 1 if yes in 'department' column, 0 if no
 
active['dairy_eggs_purchased'] = ['Yes' if x == 'dairy eggs' else 'No' for x in active['department']]
active['dairy_eggs_purchased'] = active.groupby(['user_id'])['dairy_eggs_purchased'].transform(np.max)
active['dairy_eggs_purchased'].value_counts(dropna=False)
```

### E. Create 'snacks_purchased' department profile

```python
# create column 'snacks_purchased' 1 if yes in 'department' column, 0 if no
 
active['snacks_purchased'] = ['Yes' if x == 'snacks' else 'No' for x in active['department']]
active['snacks_purchased'] = active.groupby(['user_id'])['snacks_purchased'].transform(np.max)
active['snacks_purchased'].value_counts(dropna=False)
```
