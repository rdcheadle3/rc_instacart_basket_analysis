## 03 You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.
##### In a markdown cell beneath your histogram, describe what the histogram demonstrates.

```python
# print sorted frequencies of 'order_hour_of_day'

orders_products_all['order_hour_of_day'].value_counts().sort_index()
```

```python
# create a histogram of the 'order_hour_of_day' column

orders_products_all['order_hour_of_day'].plot.hist(bins = 23)
```

```python
# save 'order_hour_of_day' bar chart as hour_hist

hour_hist = orders_products_all['order_hour_of_day'].plot.hist(bins = 23)
```

```python
# export histogram

hour_hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hour_hist.png'))
```

This histogram demonstates that sales peak between 10 AM and 5 PM. They are lowest from 12 AM to 6 AM.

## 04 The marketing team is curious about the distribution of orders among customers in terms of loyalty. Create a bar chart from the “loyalty_flag” column.

```python
# create bar chart displaying frequency of each value from 'loyalty' variable

orders_products_all['loyalty_flag'].value_counts().plot.bar()
```

```python
# create bar chart with color displaying frequency of each value from 'loyalty_flag' variable

orders_products_all['loyalty_flag'].value_counts().plot.bar(color =['blue', 'purple', 'red'])
```

```python
# save as 'loyalty flag' bar chart as loyal_bar

loyal_bar = orders_products_all['loyalty_flag'].value_counts().plot.bar(color =['blue', 'purple', 'red'])
```

```python
# export bar chart

loyal_bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'loyal_bar.png'))
```

## 05 Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day. (Hint: To check this, you need to use an accurate sample for your line chart!)

#### * no sample needed; 30% random sample applied to orders_products_all to resolve memory issues

```python
# rename 'orders_day_of_week' to 'day_of_week_order_placed'
# rename 'order_hour_of_day' to 'hour_of_day_order_placed'
# rename 'number_of_dependants' to 'number_of_dependents'

orders_products_all.rename(columns = {'orders_day_of_week' : 'day_of_week_order_placed', 'order_hour_of_day' : 'hour_of_day_order_placed', 'number_of_dependants' : 'number_of_dependents'}, inplace = True)
```

```python
# reduce dataframe orders_products_all to only include columns 'prices' and 'hour_of_day_order_placed'

hour_price = orders_products_all[['hour_of_day_order_placed','prices']]
```

```python
# create line chart of orders_products_all dataframe where
# x axis is 'hour_of_day_order_placed' and y axis is 'prices'

hour_price_line = sns.lineplot(data = hour_price, x = 'hour_of_day_order_placed',y = 'prices')
```

```python
# export hour_price_line chart

hour_price_line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hour_price_line.png'))
```

Prices of orders peak around 4-5 AM and are lowest from 9 to 11 AM.
